```julia
x = Float32[]
y = Float32[]
r = 1:1000
for i = r
    tmp = Float32(i)
    push!(x,tmp)
    push!(y,nextfloat(tmp) - tmp)
end

using Plots
scatter(x,y)
```

```julia
x = Float32(0.263789)
sum_add = 0
for i=1:10^7
    sum_add += x
end
println(sum_add)

sum = x
error = 0.0

for i=1:10^7
    y = x - error
    t = sum + y
    error = (t - sum) - y
    sum = t
end

println(sum)
```
