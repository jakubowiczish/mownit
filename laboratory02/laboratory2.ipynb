```julia
using CSV
using DataFrames
using Statistics
using Plots

input="result.csv"
myData=DataFrame()
myData=CSV.read(input, delim=",")
dfg=by(myData, :SIZE_OF_VECTOR, avg_v_time=:VECTOR_TIME => mean,std_time=:VECTOR_TIME => std,avg_m_time=:MATRIX_TIME => mean, std_matrix=:MATRIX_TIME => std)
x=dfg[:SIZE_OF_VECTOR]
y=dfg[:avg_v_time]
y2=dfg[:avg_m_time]
y_err=dfg[:std_time]

scatter(x,[y,y2],yerr=y_err,layout=2,xlabel="size",ylabel="time (ns)")


```
